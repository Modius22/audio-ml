```python
import numpy as np
import os
import pandas as pd
#from sklearn.model_selection import train_test_split
#from sklearn.tree import DecisionTreeClassifier
#from sklearn.metrics import classification_report, confusion_matrix
import librosa
import librosa.display
from pathlib import Path

```

```python

folders = ['ambulance', 'firetruck', 'traffic']

ambulance = []
firetruck = []
traffic = []


for x in range(len(folders)):
    path = '../data/train/' + folders[x]
    p = Path(path)
    for f in p.rglob('*.wav'):
        if x == 0:
            ambulance.append(f)
        if x == 1:
            firetruck.append(f)
        if x == 2:
            traffic.append(f)
```

```python

def get_features(elements,type):
    chroma_stft_list=[]
    mfcc_list=[]
    mfcc_stft_list=[]
    label_list=[]
    type_list=[]

    ind = 0
    hop_length = 5000
    for file in elements:
        try:
            filename = os.path.basename(file).rstrip(".wav")

            # 1st variable = waveform (time-series-array), 2nd = sampling rate
            y, sr = librosa.load(file)

            stft = np.abs(librosa.stft(y,hop_length=hop_length))

            #chroma_stft = np.mean(librosa.feature.chroma_stft(S=stft, sr=sr, hop_length=hop_length))
            chroma_stft = np.mean(librosa.feature.chroma_stft(S=stft, sr=sr).T,axis=0)
            chroma_stft_list.append(chroma_stft)

            #mfcc_stft = np.mean(librosa.feature.mfcc(y, S=stft, sr=sr, hop_length=hop_length))
            mfcc_stft = np.mean(librosa.feature.mfcc(y, S=stft, sr=sr,n_mfcc=40).T,axis=0)
            mfcc_stft_list.append(mfcc_stft)

            #mfcc = np.mean(librosa.feature.mfcc(y, sr=sr, hop_length=hop_length))
            mfcc = np.mean(librosa.feature.mfcc(y, sr=sr,n_mfcc=40).T,axis=0)
            mfcc_list.append(mfcc)

            #label_list.append(filename)
            type_list.append(type)

            ind += 1
        except Exception as e:
            print(e)

    #df1 = {"label":label_list}
    df2 = {"type": type_list}
    df3 = {"mean_mfcc_"+str(i):[mfcc_list[j][i] for j in range(ind)] for i in range(40)}
    df4 = {"mean_mfcc_stft_"+str(i): [mfcc_stft_list[j][i] for j in range(ind)] for i in range(40)}

    #df3 = {"chroma_stft"+str(i):[chroma_stft_list[j][i] for j in range(ind)] for i in range(200)}

    result = {}
    dicts = [ df2, df3, df4]
    for d in dicts:
        result.update(d)

    return pd.DataFrame(result)

```

```python

#0 = ambulance; 1 = firetruck, 2 = traffic
ambulance_df = get_features(ambulance, 'ambulance')
firetruck_df = get_features(firetruck, 'firetruck')
traffic_df = get_features(traffic, 'traffic')

data = pd.DataFrame()
data = data.append(ambulance_df)
data = data.append(firetruck_df)
data = data.append(traffic_df)

```

```python
data.to_csv('../data/train/data.csv')
```
